```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%pylab inline
%load_ext autoreload
%autoreload 2
```

## Scatter Matrix

### Load data from URL

```python
path = "https://raw.githubusercontent.com/cogmaster-stats/python-cogstats/master/examples/brain_size.csv"
df = pd.read_csv(path, sep=';', na_values='.', index_col=0)
display(df.head())
print("shape = {}".format(df.shape))
```

### Create a scatter matrix

- The diagnal plots are histograms
- Others are scatter plots, which can be used to check correlations.

See: https://pandas.pydata.org/docs/reference/api/pandas.plotting.scatter_matrix.html

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df[['Weight', 'Height', 'MRI_Count']], figsize=(6,6))
plt.show()
```

```python
scatter_matrix(df[['PIQ', 'VIQ', 'FSIQ']], figsize=(6,6))
plt.show()
```

```python
# Female only
df_female = df[df.Gender == 'Female']
scatter_matrix(df_female[['PIQ', 'VIQ', 'FSIQ']], figsize=(6,6))
plt.show()
```
